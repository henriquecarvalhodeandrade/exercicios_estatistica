```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 4 \n")

escore_z = 2.13

# b. Utilizar a ferramenta de distribuição normal para obter a proporção à esquerda de escore_z = 2.13
proporcao_esquerda = scipy.stats.norm.cdf(escore_z)
print(f"b. A proporção sob a curva normal padrão à esquerda de escore_z = {escore_z} é: {proporcao_esquerda:.4f}")

# a. Representação gráfica da curva (descrição e plotagem)
print("\na. Representação gráfica da curva de distribuição normal padrão e segmento sombreado à esquerda de escore_z = 2.13:")
print("Visualizemos uma curva em formato de sino com distribuição normal, centralizada em 0.")
print(f"A porção à esquerda de escore_z = {escore_z} compreende a região da curva que se estende do infinito negativo até {escore_z}, totalmente preenchida.")

# Algoritmo para gerar o gráfico (representação visual, não vinculada à "ferramenta de distribuição normal")
eixo_x = np.linspace(-4, 4, 500)
eixo_y = scipy.stats.norm.pdf(eixo_x)

plt.figure(figsize=(8, 5))
plt.plot(eixo_x, eixo_y, color='blue')
plt.fill_between(eixo_x[eixo_x < escore_z], 0, eixo_y[eixo_x < escore_z], color='skyblue', alpha=0.6)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)')
plt.axvline(escore_z, color='red', linestyle='--', label=f'escore_z = {escore_z}')
plt.title('Proporção à Esquerda de escore_z = 2.13 na Distribuição Normal Padrão')
plt.xlabel('Valor Z')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 5 \n")

escore_padronizado = -2.16

# b. Utilize a distribuição normal para identificar a proporção à esquerda de escore_padronizado = -2.16
proporcao_a_esquerda = scipy.stats.norm.cdf(escore_padronizado)
print(f"b. A proporção à esquerda de escore_padronizado = {escore_padronizado} é: {proporcao_a_esquerda:.4f}")

# c. Subtraia a proporção de 1 para obter a proporção à direita
proporcao_a_direita = 1 - proporcao_a_esquerda
print(f"c. A proporção à direita de escore_padronizado = {escore_padronizado} (1 - proporcao_a_esquerda) é: {proporcao_a_direita:.4f}")

# a. Visualização da distribuição (descrição e representação gráfica)
print("\na. Visualização da curva de distribuição normal padrão e região sombreada à direita de escore_padronizado = -2.16:")
print("Conceba uma curva em formato de sino de distribuição normal, com seu ponto central em 0.")
print(f"A região à direita de escore_padronizado = {escore_padronizado} seria a porção da curva que se estende de {escore_padronizado} até o infinito positivo, integralmente preenchida.")

# Roteiro para geração do gráfico (representação visual)
abcissa = np.linspace(-4, 4, 500)
ordenada = scipy.stats.norm.pdf(abcissa)

plt.figure(figsize=(8, 5))
plt.plot(abcissa, ordenada, color='blue')
plt.fill_between(abcissa[abcissa > escore_padronizado], 0, ordenada[abcissa > escore_padronizado], color='lightcoral', alpha=0.6)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)')
plt.axvline(escore_padronizado, color='red', linestyle='--', label=f'z = {escore_padronizado}')
plt.title('Proporção à Direita de z = -2.16 na Distribuição Normal Padrão')
plt.xlabel('Escore Z')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 6 \n")

valor_z1 = -2.165
valor_z2 = -1.35

# b. Calcule a proporção à esquerda de valor_z = -1.35 (valor_z2, o maior)
proporcao_a_esquerda_z2 = scipy.stats.norm.cdf(valor_z2)
print(f"b. A proporção à esquerda de valor_z = {valor_z2} é: {proporcao_a_esquerda_z2:.4f}")

# c. Calcule a proporção à esquerda de valor_z = -2.165 (valor_z1, o menor)
proporcao_a_esquerda_z1 = scipy.stats.norm.cdf(valor_z1)
print(f"c. A proporção à esquerda de valor_z = {valor_z1} é: {proporcao_a_esquerda_z1:.4f}")

# d. Subtraia a proporção menor da proporção maior
# Proporção entre valor_z1 e valor_z2 = (Proporção à esquerda de valor_z2) - (Proporção à esquerda de valor_z1)
proporcao_entre = proporcao_a_esquerda_z2 - proporcao_a_esquerda_z1
print(f"d. A proporção entre valor_z = {valor_z1} e valor_z = {valor_z2} é (Proporção à esquerda de {valor_z2}) - (Proporção à esquerda de {valor_z1}) = {proporcao_entre:.4f}")

# a. Representação visual da curva (descrição e plotagem)
print("\na. Representação visual da curva normal padrão e região sombreada entre valor_z = -2.165 e valor_z = -1.35:")
print("Visualize uma curva de sino de distribuição normal com o ponto central em 0.")
print(f"A área entre valor_z = {valor_z1} e valor_z = {valor_z2} seria a parte da curva contida entre esses dois escores, totalmente preenchida.")

# Algoritmo para gerar o gráfico (visualização)
eixo_horizontal = np.linspace(-4, 4, 500)
eixo_vertical = scipy.stats.norm.pdf(eixo_horizontal)

plt.figure(figsize=(8, 5))
plt.plot(eixo_horizontal, eixo_vertical, color='blue')
# Sombrear a área entre valor_z1 e valor_z2
eixo_preenchimento = np.linspace(valor_z1, valor_z2, 100)
eixo_vertical_preenchimento = scipy.stats.norm.pdf(eixo_preenchimento)
plt.fill_between(eixo_preenchimento, 0, eixo_vertical_preenchimento, color='lightgreen', alpha=0.6)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)')
plt.axvline(valor_z1, color='red', linestyle='--', label=f'z1 = {valor_z1}')
plt.axvline(valor_z2, color='purple', linestyle='--', label=f'z2 = {valor_z2}')
plt.title(f'Proporção entre z = {valor_z1} e z = {valor_z2} na Distribuição Normal Padrão')
plt.xlabel('Escore Z')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# e. Analise os resultados
print("\ne. Análise dos resultados:")
print(f"A proporção calculada entre valor_z = {valor_z1} e valor_z = {valor_z2} é de {proporcao_entre:.4f}, ou {proporcao_entre*100:.2f}%.")
print("Isso indica que aproximadamente essa percentagem dos dados em uma distribuição normal padrão estaria compreendida entre esses dois valores de escore-z.")
print("Considerando que ambos os escores-z são negativos, a região está localizada no lado esquerdo (abaixo da média) da distribuição.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 7 \n")

# Informações fornecidas
valor_medio = 67  # milhas por hora
desvio_padrao_calculado = 3.5  # milhas por hora
velocidade_limite = 70  # milhas por hora

# a. Represente graficamente.
print("\na. Representação gráfica da distribuição normal:")
print("Conceba uma curva em formato de sino (distribuição normal) centralizada em 67 milhas/hora.")
print("A porção que buscamos localizar é a área da curva situada à direita de 70 milhas/hora.")

# Configuração para o plot
# Gerar valores para o eixo horizontal da curva, cobrindo cerca de 4 desvios padrão para cada lado da média
eixo_x = np.linspace(valor_medio - 4 * desvio_padrao_calculado, valor_medio + 4 * desvio_padrao_calculado, 500)
# Calcular os valores do eixo vertical da função de densidade de probabilidade (FDP) para a curva normal
eixo_y = scipy.stats.norm.pdf(eixo_x, loc=valor_medio, scale=desvio_padrao_calculado)

plt.figure(figsize=(10, 6))  # Define o tamanho da figura do gráfico
plt.plot(eixo_x, eixo_y, color='blue', label='Distribuição de Desempenho')  # Plota a curva normal
plt.axvline(valor_medio, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {valor_medio} mph')  # Adiciona linha para a média
plt.axvline(velocidade_limite, color='red', linestyle='--', label=f'Limite de Velocidade = {velocidade_limite} mph')  # Adiciona linha para o limite de velocidade

# Preencher a área à direita do limite de velocidade
# Cria um novo array para o eixo horizontal apenas para a área a ser preenchida
eixo_x_preenchimento = np.linspace(velocidade_limite, valor_medio + 4 * desvio_padrao_calculado, 100)
# Calcula os valores do eixo vertical para a área preenchida
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=valor_medio, scale=desvio_padrao_calculado)
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='salmon', alpha=0.6, label='Probabilidade > 70 mph')  # Preenche a área

plt.title('Distribuição de Desempenho de Veículos')  # Título do gráfico
plt.xlabel('Velocidade (milhas por hora)')  # Rótulo do eixo X
plt.ylabel('Densidade de Probabilidade')  # Rótulo do eixo Y
plt.legend()  # Mostra a legenda
plt.grid(True, linestyle=':', alpha=0.7)  # Adiciona grade
plt.show()  # Exibe o gráfico


# b. Determine o escore-padrão correspondente a 70 milhas por hora.
escore_padrao = (velocidade_limite - valor_medio) / desvio_padrao_calculado
print(f"\nb. O escore-padrão correspondente a {velocidade_limite} milhas por hora é: {escore_padrao:.2f}")

# c. Encontre a proporção à direita do escore-padrão.
# A função cdf() (função de distribuição acumulada) retorna a proporção à esquerda.
# Para a proporção à direita, subtraímos de 1.
proporcao_esquerda = scipy.stats.norm.cdf(escore_padrao)
proporcao_direita = 1 - proporcao_esquerda
print(f"c. A proporção à direita do escore-padrão ({escore_padrao:.2f}) é: {proporcao_direita:.4f}")

# d. Analise os resultados.
print("\nd. Análise dos resultados:")
print(f"A probabilidade de um veículo selecionado ao acaso estar superando o limite de velocidade de {velocidade_limite} milhas por hora é de aproximadamente {proporcao_direita:.4f}.")
print(f"Isso significa que cerca de {proporcao_direita*100:.2f}% dos veículos estão se deslocando a mais de {velocidade_limite} milhas por hora.")
```

```python
import scipy.stats  # Para funções da distribuição normal (cdf, pdf)
import matplotlib.pyplot as plt  # Para plotar diagramas
import numpy as np  # Para operações numéricas, como criar intervalos para o plot

print(" Exercício 8 \n")

# Dados fornecidos do problema
media_tempo_geral = 45  # média de duração em minutos
desvio_padrao_duracao = 12  # desvio padrão em minutos
limite_inferior_tempo = 33  # limiar inferior de duração em minutos
limite_superior_tempo = 60  # limiar superior de duração em minutos

# a. Delineie um diagrama da distribuição normal
print("\na. Delineamento do diagrama da distribuição normal:")
print("Conceba uma curva em formato de sino (distribuição normal) centralizada em 45 minutos.")
print(f"A porção que buscamos identificar é a área da curva entre {limite_inferior_tempo} e {limite_superior_tempo} minutos.")

# Gerar valores para o eixo horizontal da curva normal. Empregamos um intervalo de aproximadamente 4 desvios padrão
# para cada lado da média para abranger a maior parte da distribuição.
eixo_x = np.linspace(media_tempo_geral - 4 * desvio_padrao_duracao, media_tempo_geral + 4 * desvio_padrao_duracao, 500)
# Calcular os valores do eixo vertical da função de densidade de probabilidade (FDP) para a curva normal.
# 'loc' indica a média e 'scale' o desvio padrão.
eixo_y = scipy.stats.norm.pdf(eixo_x, loc=media_tempo_geral, scale=desvio_padrao_duracao)

plt.figure(figsize=(10, 6))  # Ajusta as dimensões da figura do diagrama para melhor visualização
plt.plot(eixo_x, eixo_y, color='blue', label='Distribuição do Período na Loja')  # Traça a curva normal

# Inserir linhas verticais para a média e os limiares de duração
plt.axvline(media_tempo_geral, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_tempo_geral} min')
plt.axvline(limite_inferior_tempo, color='red', linestyle='--', label=f'Limiar Inferior = {limite_inferior_tempo} min')
plt.axvline(limite_superior_tempo, color='green', linestyle='--', label=f'Limiar Superior = {limite_superior_tempo} min')

# Preencher a porção de interesse (entre limite_inferior_tempo e limite_superior_tempo)
# Criar um arranjo eixo_x_preenchimento apenas para a seção que será preenchida
eixo_x_preenchimento = np.linspace(limite_inferior_tempo, limite_superior_tempo, 100)
# Calcular os valores do eixo vertical correspondentes para essa seção
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_tempo_geral, scale=desvio_padrao_duracao)
# Preencher a porção entre a curva e o eixo horizontal, de 0 até eixo_y_preenchimento
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='lightblue', alpha=0.6, label=f'Probabilidade entre {limite_inferior_tempo} e {limite_superior_tempo} min')

plt.title('Distribuição da Duração de Permanência no Supermercado')  # Título do diagrama
plt.xlabel('Duração (minutos)')  # Rótulo do eixo horizontal
plt.ylabel('Densidade de Probabilidade')  # Rótulo do eixo vertical
plt.legend()  # Exibe a legenda para identificar os elementos do diagrama
plt.grid(True, linestyle=':', alpha=0.7)  # Adiciona uma grade ao diagrama
plt.show()  # Apresenta o diagrama


# b. Calcule os escores-z que correspondem a 33 minutos e a 60 minutos.
# Fórmula do escore-z: Z = (x - média) / desvio_padrao
escore_z_min = (limite_inferior_tempo - media_tempo_geral) / desvio_padrao_duracao
escore_z_max = (limite_superior_tempo - media_tempo_geral) / desvio_padrao_duracao
print(f"\nb. O escore-z para {limite_inferior_tempo} minutos é: {escore_z_min:.2f}")
print(f"O escore-z para {limite_superior_tempo} minutos é: {escore_z_max:.2f}")

# c. Encontre a proporção acumulada para cada escore-z e subtraia a proporção menor da proporção maior.
# scipy.stats.norm.cdf(z) retorna a proporção acumulada à esquerda do escore-z.
proporcao_acumulada_escore_min = scipy.stats.norm.cdf(escore_z_min)
proporcao_acumulada_escore_max = scipy.stats.norm.cdf(escore_z_max)

# A probabilidade entre dois escores-z é a diferença das proporções acumuladas.
chance_entre_duracoes = proporcao_acumulada_escore_max - proporcao_acumulada_escore_min
print(f"\nc. A proporção acumulada para z = {escore_z_min:.2f} é: {proporcao_acumulada_escore_min:.4f}")
print(f"A proporção acumulada para z = {escore_z_max:.2f} é: {proporcao_acumulada_escore_max:.4f}")
print(f"A probabilidade de um consumidor permanecer entre {limite_inferior_tempo} e {limite_superior_tempo} minutos é: {chance_entre_duracoes:.4f}")

# d. Explique sua resposta quando 150 consumidores entram na loja.
# Quantos consumidores você estimaria que permaneceriam na loja entre 33 e 60 minutos?
quantidade_consumidores = 150
# O número estimado é a probabilidade multiplicada pelo total de consumidores.
consumidores_estimados = chance_entre_duracoes * quantidade_consumidores

print("\nd. Explicação dos resultados:")
print(f"A probabilidade de um único consumidor permanecer na loja entre {limite_inferior_tempo} e {limite_superior_tempo} minutos é de aproximadamente {chance_entre_duracoes*100:.2f}%.")
print(f"Quando {quantidade_consumidores} consumidores acessam a loja, estimamos que aproximadamente {consumidores_estimados:.0f} consumidores (arredondado) permaneçam na loja entre {limite_inferior_tempo} e {limite_superior_tempo} minutos.")
print(f"O valor preciso é {consumidores_estimados:.2f} consumidores.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 9 \n")

# Dados obtidos da pesquisa (exemplo, baseados em valores típicos encontrados)
# Média do nível de triglicerídeos na população dos EUA (ex: de 2017-2018)
# Fonte: "US Trends in Cholesterol Screening, Lipid Levels, and Lipid‐Lowering Medication Use in US Adults, 1999 to 2018"
media_niveis_lipidicos = 91.4  # mg/dL

# Estimação do desvio padrão com base no IQR de uma das fontes (ex: IQR de 46 de uma das tabelas: 114 - 68)
# Para uma distribuição normal, IQR approx 1.34 * sigma. Então sigma = IQR / 1.34
desvio_padrao_lipidios = 46 / 1.34  # mg/dL
# Arredondando para 2 casas decimais para clareza
desvio_padrao_lipidios = round(desvio_padrao_lipidios, 2)

# Limites dos triglicerídeos de interesse
patamar_inferior = 100  # mg/dL
patamar_superior = 150  # mg/dL

print(f"Assumindo uma distribuição normal com: ")
print(f"Média ($\mu$) = {media_niveis_lipidicos} mg/dL")
print(f"Desvio Padrão ($\sigma$) = {desvio_padrao_lipidios} mg/dL")

# a. Representação gráfica.
print("\na. Representação gráfica da distribuição normal:")
print(f"A curva em formato de sino demonstra a distribuição dos níveis de triglicerídeos, com centro em {media_niveis_lipidicos} mg/dL.")
print(f"A região que desejamos determinar é a porção da curva entre {patamar_inferior} e {patamar_superior} mg/dL.")

# Configuração para o diagrama
# Gerar valores para o eixo horizontal da curva, cobrindo cerca de 4 desvios padrão para cada lado da média
eixo_x = np.linspace(media_niveis_lipidicos - 4 * desvio_padrao_lipidios, media_niveis_lipidicos + 4 * desvio_padrao_lipidios, 500)
# Calcular os valores do eixo vertical da função de densidade de probabilidade (FDP)
eixo_y = scipy.stats.norm.pdf(eixo_x, loc=media_niveis_lipidicos, scale=desvio_padrao_lipidios)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, color='blue', label='Distribuição do Nível de Triglicerídeos')
plt.axvline(media_niveis_lipidicos, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_niveis_lipidicos:.1f} mg/dL')
plt.axvline(patamar_inferior, color='red', linestyle='--', label=f'Patamar Inferior = {patamar_inferior} mg/dL')
plt.axvline(patamar_superior, color='green', linestyle='--', label=f'Patamar Superior = {patamar_superior} mg/dL')

# Preencher a área entre patamar_inferior e patamar_superior
eixo_x_preenchimento = np.linspace(patamar_inferior, patamar_superior, 100)
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_niveis_lipidicos, scale=desvio_padrao_lipidios)
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='lightgreen', alpha=0.6, label=f'Probabilidade entre {patamar_inferior} e {patamar_superior} mg/dL')

plt.title('Distribuição dos Níveis de Triglicerídeos nos EUA')
plt.xlabel('Nível de Triglicerídeos (mg/dL)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# b. Determine os escores-padrão que correspondem a 100 e 150 mg/dL.
escore_padrao_inferior = (patamar_inferior - media_niveis_lipidicos) / desvio_padrao_lipidios
escore_padrao_superior = (patamar_superior - media_niveis_lipidicos) / desvio_padrao_lipidios

print(f"\nb. O escore-padrão para {patamar_inferior} mg/dL é: {escore_padrao_inferior:.2f}")
print(f"O escore-padrão para {patamar_superior} mg/dL é: {escore_padrao_superior:.2f}")

# c. Encontre a proporção acumulada para cada escore-padrão e subtraia a proporção menor da proporção maior.
proporcao_acumulada_escore_inferior = scipy.stats.norm.cdf(escore_padrao_inferior)
proporcao_acumulada_escore_superior = scipy.stats.norm.cdf(escore_padrao_superior)

chance_entre_limites = proporcao_acumulada_escore_superior - proporcao_acumulada_escore_inferior

print(f"\nc. A proporção acumulada para z = {escore_padrao_inferior:.2f} é: {proporcao_acumulada_escore_inferior:.4f}")
print(f"A proporção acumulada para z = {escore_padrao_superior:.2f} é: {proporcao_acumulada_escore_superior:.4f}")
print(f"A probabilidade de o nível de triglicerídeo estar entre {patamar_inferior} e {patamar_superior} mg/dL é: {chance_entre_limites:.4f}")

# d. Analise os resultados.
print("\nd. Análise dos resultados:")
print(f"A probabilidade de um indivíduo escolhido ao acaso nos Estados Unidos ter um nível de triglicerídeo entre {patamar_inferior} e {patamar_superior} mg/dL é de aproximadamente {chance_entre_limites:.4f}, ou {chance_entre_limites*100:.2f}%.")
print(f"Isso indica que cerca de {chance_entre_limites*100:.2f}% da população possui níveis de triglicerídeos dentro dessa faixa, conforme as estimativas de média e desvio padrão utilizadas.")
```

```python
import scipy.stats

print(" Exercício 10.2 \n")

proporcao_central = 0.95
proporcao_nas_extremidades = 1 - proporcao_central
proporcao_uma_extremidade = proporcao_nas_extremidades / 2

# a. Determine a proporção acumulada.
proporcao_acumulada_valor_z_negativo = proporcao_uma_extremidade
proporcao_acumulada_valor_z_positivo = proporcao_central + proporcao_uma_extremidade

print(f"a. A proporção acumulada à esquerda de -valor_z é: {proporcao_acumulada_valor_z_negativo:.4f}")
print(f"   A proporção acumulada à esquerda de +valor_z é: {proporcao_acumulada_valor_z_positivo:.4f}")

# b. Localize a proporção na ferramenta de distribuição normal. (Feito pela função ppf)
# c. Encontre o valor-z que corresponde à proporção.
valor_z_negativo = scipy.stats.norm.ppf(proporcao_acumulada_valor_z_negativo)
valor_z_positivo = scipy.stats.norm.ppf(proporcao_acumulada_valor_z_positivo)

print(f"c. O valor-z negativo é: {valor_z_negativo:.2f}")
print(f"   O valor-z positivo é: {valor_z_positivo:.2f}")
print(f"Portanto, o valor-z para o qual 95% da proporção da distribuição esteja entre -valor_z e +valor_z é aproximadamente +/- {valor_z_positivo:.2f}.")
```

```python
print(" Exercício 12 \n")

# a. Identifique a média e o desvio padrão da distribuição normal.
media_peso_animal = 52       # libras (µ)
desvio_padrao_animal = 15    # libras (s)

print(f"a. Média (μ) = {media_peso_animal} libras, Desvio Padrão (σ) = {desvio_padrao_animal} libras.")

valores_z_fornecidos = [-2.33, 3.10, 0.58]
pesos_resultantes = []

print("\nb. Converta cada escore-z em um valor de peso (x):")
for z_valor in valores_z_fornecidos:
    peso_real = media_peso_animal + z_valor * desvio_padrao_animal
    pesos_resultantes.append((z_valor, peso_real))
    print(f"Para z = {z_valor}, o peso x é: {peso_real:.2f} libras")

print("\nc. Analise os resultados:")
for z_valor, peso_real in pesos_resultantes:
    if z_valor < 0:
        print(f"Um cão com z = {z_valor} possui um peso de {peso_real:.2f} libras, o que está {abs(z_valor):.2f} desvios padrão ABAIXO da média.")
    elif z_valor > 0:
        print(f"Um cão com z = {z_valor} possui um peso de {peso_real:.2f} libras, o que está {z_valor:.2f} desvios padrão ACIMA da média.")
    else:
        print(f"Um cão com z = {z_valor} possui um peso de {peso_real:.2f} libras, o que corresponde EXATAMENTE à média.")

print("\nAnálise geral:")
print("Cães com escores-z negativos apresentam pesos inferiores à média, enquanto cães com escores-z positivos têm pesos superiores à média. Quanto maior o valor absoluto do escore-z, mais distante da média o peso se encontra.")
print("Por exemplo, um cão com z = -2.33 (2.33 desvios padrão abaixo da média) pesa 17.05 libras, sendo consideravelmente mais leve que a média.")
print("Um cão com z = 3.10 (3.10 desvios padrão acima da média) pesa 98.50 libras, sendo consideravelmente mais pesado que a média.")
print("Um cão com z = 0.58 (0.58 desvios padrão acima da média) pesa 60.70 libras, sendo ligeiramente mais pesado que a média.")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print(" Exercício 13 \n")

# Informações providenciadas
media_parada = 129        # pés (µ)
desvio_padrao_parada = 5.18 # pés (σ)
percentil_requerido = 0.01 # 1% inferior

# a. Delineie um diagrama.
print("\na. Delineamento do diagrama da distribuição normal:")
print(f"A curva ilustra a distribuição das distâncias de parada, centralizada em {media_parada} pés.")
print(f"Buscamos identificar a distância de parada que demarca o limite inferior de 1% da distribuição.")

# Configuração para o diagrama
eixo_x = np.linspace(media_parada - 4 * desvio_padrao_parada, media_parada + 4 * desvio_padrao_parada, 500)
eixo_y = scipy.stats.norm.pdf(eixo_x, loc=media_parada, scale=desvio_padrao_parada)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x, eixo_y, color='blue', label='Distribuição de Distâncias de Parada')
plt.axvline(media_parada, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_parada} pés')

# b. Ache o escore-z que corresponda à área fornecida.
escore_z_percentil = scipy.stats.norm.ppf(percentil_requerido)
print(f"\nb. O escore-z que corresponde ao percentil {percentil_requerido*100:.0f}% é: {escore_z_percentil:.2f}")

# c. Calcule o valor_x utilizando a fórmula valor_x = µ + zσ.
distancia_parada_limite = media_parada + escore_z_percentil * desvio_padrao_parada
print(f"c. A distância de parada correspondente é: {distancia_parada_limite:.2f} pés")

# Adicionar linha vertical para o limite calculado e preencher a região
plt.axvline(distancia_parada_limite, color='red', linestyle='--', label=f'Limite Inferior de 1% = {distancia_parada_limite:.2f} pés')
eixo_x_preenchimento = np.linspace(media_parada - 4 * desvio_padrao_parada, distancia_parada_limite, 100)
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_parada, scale=desvio_padrao_parada)
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='purple', alpha=0.6, label=f'1% mais baixo')

plt.title('Distribuição das Distâncias de Parada')
plt.xlabel('Distância de Parada (pés)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# d. Explique o resultado.
print("\nd. Explicação do resultado:")
print(f"A maior distância de parada que um desses veículos poderia apresentar e ainda pertencer ao grupo do 1% inferior é de {distancia_parada_limite:.2f} pés.")
print("Isso significa que 1% dos carros testados teriam uma distância de parada igual ou inferior a 116.03 pés (com base nos dados fornecidos), indicando uma performance de frenagem excepcionalmente boa (menor distância = melhor).")
```

```python
import scipy.stats  # Módulo para funções estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt  # Módulo para construção de visualizações gráficas
import numpy as np  # Módulo para operações numéricas, como geração de arranjos

print(" Exercício 14 \n")

# Dados fornecidos no contexto do problema
media_periodo_servico = 11.2  # Média do período de trabalho dos colaboradores (em anos)
desvio_padrao_periodo = 2.1  # Desvio padrão do período de trabalho (em anos)
percentil_limite = 0.10  # Os 10% com menor tempo na organização (representa a porção à esquerda)

# a. Delineie um diagrama da distribuição normal.
print("\na. Delineamento do diagrama da distribuição normal:")
print(f"A curva ilustra a distribuição do período de serviço dos colaboradores, centralizada em {media_periodo_servico} anos.")
print(f"Buscamos determinar o tempo máximo que define o limiar inferior de {percentil_limite*100:.0f}% da distribuição.")

# Gerar valores para o eixo horizontal do diagrama.
# Empregamos um intervalo de aproximadamente 4 desvios padrão para cada lado da média
# para abranger a maior parte da distribuição normal.
eixo_x = np.linspace(media_periodo_servico - 4 * desvio_padrao_periodo, media_periodo_servico + 4 * desvio_padrao_periodo, 500)
# Calcular os valores do eixo vertical da Função de Densidade de Probabilidade (FDP)
# da distribuição normal para cada valor de x. 'loc' indica a média e 'scale' o desvio padrão.
eixo_y = scipy.stats.norm.pdf(eixo_x, loc=media_periodo_servico, scale=desvio_padrao_periodo)

plt.figure(figsize=(10, 6))  # Ajusta as dimensões da figura para melhor visibilidade
plt.plot(eixo_x, eixo_y, color='blue', label='Distribuição do Período de Serviço')  # Traça a curva da distribuição

# Inserir uma linha vertical para a média da distribuição
plt.axvline(media_periodo_servico, color='gray', linestyle='--', linewidth=0.8, label=f'Média = {media_periodo_servico} anos')

# b. Apure o escore-z que corresponda à porção informada.
# A função norm.ppf() (Função de Ponto Percentil ou função inversa da FDC)
# retorna o valor do escore-z para uma determinada porção acumulada à esquerda.
escore_z_limite = scipy.stats.norm.ppf(percentil_limite)
print(f"\nb. O escore-z que corresponde ao percentil {percentil_limite*100:.0f}% é: {escore_z_limite:.2f}")

# c. Calcule o valor_x utilizando a fórmula valor_x = µ + zσ.
# Esta fórmula converte um escore-z de volta para a escala original dos dados (tempo em anos).
periodo_limite_corte = media_periodo_servico + escore_z_limite * desvio_padrao_periodo
print(f"c. O período máximo de trabalho para ser elegível ao corte é: {periodo_limite_corte:.2f} anos")

# Inserir uma linha vertical no diagrama para o período máximo de corte calculado
plt.axvline(periodo_limite_corte, color='red', linestyle='--', label=f'Limiar de Corte (10%) = {periodo_limite_corte:.2f} anos')

# Preencher a porção no diagrama que representa os 10% com menor tempo na organização
# Gera um novo arranjo eixo_x_preenchimento apenas para a porção que será preenchida (do extremo esquerdo até o limiar de corte)
eixo_x_preenchimento = np.linspace(media_periodo_servico - 4 * desvio_padrao_periodo, periodo_limite_corte, 100)
# Calcula os valores do eixo vertical correspondentes para esta seção da curva
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_periodo_servico, scale=desvio_padrao_periodo)
# Preenche a porção entre a curva e o eixo horizontal
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='purple', alpha=0.6, label=f'{percentil_limite*100:.0f}% com menor tempo')

plt.title('Distribuição do Período de Serviço e Limiar de Corte')  # Título do diagrama
plt.xlabel('Período de Serviço (anos)')  # Rótulo do eixo horizontal
plt.ylabel('Densidade de Probabilidade')  # Rótulo do eixo vertical
plt.legend()  # Exibe a legenda do diagrama
plt.grid(True, linestyle=':', alpha=0.7)  # Adiciona uma grade
plt.show()  # Apresenta o diagrama

# d. Explique o resultado.
print("\nd. Explicação do resultado:")
print(f"O período máximo que um colaborador pode ter atuado na organização e ainda assim ser incluído no corte (pertencer aos 10% com menor tempo) é de {periodo_limite_corte:.2f} anos.")
print(f"Isso significa que qualquer colaborador com um período de serviço igual ou inferior a {periodo_limite_corte:.2f} anos (com base nas informações fornecidas) seria desvinculado nesta redução de quadro.")
```

```python
import scipy.stats  # Módulo para funcionalidades estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt  # Módulo para a criação de visualizações gráficas
import numpy as np  # Módulo para operações numéricas, como a geração de arranjos

print(" Exercício 15 \n")

# Informações fornecidas no enunciado do problema
media_conjunto_total = 47     # Média das despesas de telefone celular (µ)
desvio_padrao_conjunto_total = 9  # Desvio padrão das despesas de telefone celular (σ)
dimensao_subgrupo = 64      # Número de despesas de telefone celular em cada subgrupo (n)

# 1. Calcule a média da distribuição dos subgrupos de médias
media_distribuicao_subgrupos = media_conjunto_total
print(f"A média da distribuição dos subgrupos de médias (µ_x̄) é: US$ {media_distribuicao_subgrupos:.2f}")

# 2. Calcule o desvio padrão da distribuição dos subgrupos de médias (Erro Padrão)
desvio_padrao_distribuicao_subgrupos = desvio_padrao_conjunto_total / np.sqrt(dimensao_subgrupo)
print(f"O desvio padrão da distribuição dos subgrupos de médias (σ_x̄) é: US$ {desvio_padrao_distribuicao_subgrupos:.2f}")

# 3. Delineie um diagrama da distribuição dos subgrupos.
print("\nDelineamento do diagrama da distribuição dos subgrupos de médias:")
print("A distribuição dos subgrupos de médias é também uma curva normal, mas é mais compacta (apresenta um desvio padrão inferior) do que a distribuição do conjunto total original, em virtude do Teorema do Limite Central.")

# Configuração para o diagrama
# Defina um intervalo para o eixo horizontal do diagrama (aproximadamente 4 desvios padrão em torno da média)
# Tanto para a distribuição do conjunto total quanto para a distribuição dos subgrupos para comparação
eixo_x_conjunto_total = np.linspace(media_conjunto_total - 4 * desvio_padrao_conjunto_total, media_conjunto_total + 4 * desvio_padrao_conjunto_total, 500)
eixo_y_conjunto_total = scipy.stats.norm.pdf(eixo_x_conjunto_total, loc=media_conjunto_total, scale=desvio_padrao_conjunto_total)

eixo_x_subgrupos = np.linspace(media_distribuicao_subgrupos - 4 * desvio_padrao_distribuicao_subgrupos, media_distribuicao_subgrupos + 4 * desvio_padrao_distribuicao_subgrupos, 500)
eixo_y_subgrupos = scipy.stats.norm.pdf(eixo_x_subgrupos, loc=media_distribuicao_subgrupos, scale=desvio_padrao_distribuicao_subgrupos)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x_conjunto_total, eixo_y_conjunto_total, color='skyblue', label='Distribuição do Conjunto Total Original (σ=9)', linestyle='--', alpha=0.7)
plt.plot(eixo_x_subgrupos, eixo_y_subgrupos, color='blue', label=f'Distribuição dos Subgrupos de Médias (n={dimensao_subgrupo}, σ_x̄={desvio_padrao_distribuicao_subgrupos:.2f})')

plt.axvline(media_conjunto_total, color='gray', linestyle=':', linewidth=0.8, label=f'Média (µ = µ_x̄) = {media_conjunto_total} US$')

plt.title('Comparativo da Distribuição do Conjunto Total e da Distribuição dos Subgrupos de Médias')
plt.xlabel('Despesas de Telefone Celular (US$)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# Análise
print("\nAnálise:")
print(f"A média das médias dos subgrupos (US$ {media_distribuicao_subgrupos:.2f}) é idêntica à média do conjunto total original.")
print(f"O desvio padrão da distribuição dos subgrupos de médias (erro padrão = US$ {desvio_padrao_distribuicao_subgrupos:.2f}) é consideravelmente menor que o desvio padrão do conjunto total (US$ {desvio_padrao_conjunto_total}),")
print("indicando que as médias dos subgrupos são menos dispersas e tendem a se agrupar mais próximas da média do conjunto total.")
print("O diagrama demonstra que a distribuição dos subgrupos é mais elevada e estreita do que a distribuição do conjunto total, o que reflete sua menor variabilidade.")
```

```python
import scipy.stats  # Módulo para funcionalidades estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt  # Módulo para a criação de visualizações gráficas
import numpy as np  # Módulo para operações numéricas, como a geração de arranjos

print(" Exercício 16 \n")

# Informações fornecidas
media_universo = 3.5     # Média dos diâmetros dos carvalhos brancos (µ)
desvio_padrao_universo = 0.2 # Desvio padrão dos diâmetros (σ)
tamanho_da_selecao = 16    # Número de carvalhos brancos em cada seleção (n)

# a. Encontre a média e o desvio padrão da distribuição de seleção das médias.

# Média da distribuição de seleção das médias (µx-barra)
media_distribuicao_de_selecao = media_universo
print(f"a. A média da distribuição de seleção das médias (µ_x̄) é: {media_distribuicao_de_selecao:.2f} pés")

# Desvio padrão da distribuição de seleção das médias (σx-barra ou Erro Padrão)
desvio_padrao_distribuicao_de_selecao = desvio_padrao_universo / np.sqrt(tamanho_da_selecao)
print(f"   O desvio padrão da distribuição de seleção das médias (σ_x̄) é: {desvio_padrao_distribuicao_de_selecao:.3f} pés")

# b. Delineie uma curva normal com média da distribuição de seleção e desvio padrão da distribuição de seleção.
print("\nb. Delineamento de uma curva normal para a distribuição de seleção das médias:")
print("A distribuição de seleção das médias é uma curva normal mais compacta e elevada, centralizada na mesma média do universo original.")

# Configuração para o diagrama
# Defina um intervalo para o eixo horizontal do diagrama, cobrindo aproximadamente 4 desvios padrão em torno da média
# Isso será feito para a distribuição do universo original e para a distribuição de seleção para comparação.
eixo_x_universo = np.linspace(media_universo - 4 * desvio_padrao_universo, media_universo + 4 * desvio_padrao_universo, 500)
eixo_y_universo = scipy.stats.norm.pdf(eixo_x_universo, loc=media_universo, scale=desvio_padrao_universo)

eixo_x_selecao = np.linspace(media_distribuicao_de_selecao - 4 * desvio_padrao_distribuicao_de_selecao, media_distribuicao_de_selecao + 4 * desvio_padrao_distribuicao_de_selecao, 500)
eixo_y_selecao = scipy.stats.norm.pdf(eixo_x_selecao, loc=media_distribuicao_de_selecao, scale=desvio_padrao_distribuicao_de_selecao)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x_universo, eixo_y_universo, color='skyblue', label=f'Distribuição do Universo Original (σ={desvio_padrao_universo:.1f})', linestyle='--', alpha=0.7)
plt.plot(eixo_x_selecao, eixo_y_selecao, color='blue', label=f'Distribuição de Seleção das Médias (n={tamanho_da_selecao}, σ_x̄={desvio_padrao_distribuicao_de_selecao:.3f})')

plt.axvline(media_universo, color='gray', linestyle=':', linewidth=0.8, label=f'Média (µ = µ_x̄) = {media_universo} pés')

plt.title('Comparativo da Distribuição do Diâmetro dos Carvalhos e da Distribuição de Seleção das Médias')
plt.xlabel('Diâmetro (pés)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# Análise
print("\nAnálise:")
print(f"A média das médias das seleções ({media_distribuicao_de_selecao:.2f} pés) é idêntica à média do universo de carvalhos.")
print(f"O desvio padrão da distribuição de seleção das médias (erro padrão = {desvio_padrao_distribuicao_de_selecao:.3f} pés) é substancialmente inferior ao desvio padrão do universo ({desvio_padrao_universo:.1f} pés).")
print("Isso demonstra que as médias de seleções de 16 carvalhos tendem a se agrupar muito mais próximas da verdadeira média do universo do que os diâmetros individuais.")
print("O diagrama visualiza essa menor variabilidade, exibindo uma curva mais elevada e compacta para a distribuição de seleção em contraste com a distribuição do universo original.")
```

```python
import scipy.stats  # Módulo para funcionalidades estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt  # Módulo para a criação de visualizações gráficas
import numpy as np  # Módulo para operações numéricas, como a geração de arranjos

print(" Exercício 17 \n")

# Informações fornecidas
media_referencia = 25     # minutos (µ) para 15-19 anos
desvio_padrao_referencia = 1.5  # minutos (σ)
tamanho_do_grupo = 100    # número de condutores no grupo (n)

# Limites do período médio para a probabilidade
periodo_medio_inferior = 24.7  # minutos
periodo_medio_superior = 25.5  # minutos

# a. Empregue o teorema do limite central para encontrar a média e o desvio padrão da distribuição de grupos de médias.
media_distribuicao_de_grupos = media_referencia
desvio_padrao_distribuicao_de_grupos = desvio_padrao_referencia / np.sqrt(tamanho_do_grupo)

print(f"a. Média da distribuição de grupos de médias (µ_x̄) = {media_distribuicao_de_grupos:.2f} minutos")
print(f"   Desvio padrão da distribuição de grupos de médias (σ_x̄) = {desvio_padrao_distribuicao_de_grupos:.3f} minutos")

# Delineie a distribuição de grupos de médias.
print("\nDelineamento do diagrama da distribuição de grupos de médias:")
print("A distribuição de grupos de médias é uma curva normal, centrada na média da população, e é mais compacta devido ao Teorema do Limite Central.")

# Configuração para o diagrama
# Intervalo para o eixo horizontal da distribuição de grupos (cobrir +/- 4 desvios padrão)
eixo_x_grupos = np.linspace(media_distribuicao_de_grupos - 4 * desvio_padrao_distribuicao_de_grupos,
                            media_distribuicao_de_grupos + 4 * desvio_padrao_distribuicao_de_grupos, 500)
eixo_y_grupos = scipy.stats.norm.pdf(eixo_x_grupos, loc=media_distribuicao_de_grupos, scale=desvio_padrao_distribuicao_de_grupos)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x_grupos, eixo_y_grupos, color='blue', label=f'Distribuição de Grupos (n={tamanho_do_grupo}, σ_x̄={desvio_padrao_distribuicao_de_grupos:.3f})')
plt.axvline(media_distribuicao_de_grupos, color='gray', linestyle='--', linewidth=0.8, label=f'Média (µ_x̄) = {media_distribuicao_de_grupos} min')

# Adicionar linhas para os limites de 24.7 e 25.5
plt.axvline(periodo_medio_inferior, color='red', linestyle='--', label=f'x̄ = {periodo_medio_inferior} min')
plt.axvline(periodo_medio_superior, color='green', linestyle='--', label=f'x̄ = {periodo_medio_superior} min')

# Preencher a porção entre os limites
eixo_x_preenchimento = np.linspace(periodo_medio_inferior, periodo_medio_superior, 100)
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_distribuicao_de_grupos, scale=desvio_padrao_distribuicao_de_grupos)
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='lightblue', alpha=0.6, label=f'Probabilidade entre {periodo_medio_inferior} e {periodo_medio_superior} min')

plt.title(f'Distribuição de Grupos de Médias para Condutores de 15-19 anos (n={tamanho_do_grupo})')
plt.xlabel('Período Médio Dispêndido Conduzindo (minutos)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

# b. Calcule os escores-z que correspondam a x-barra = 24,7 minutos e x-barra = 25,5 minutos.
z_inferior = (periodo_medio_inferior - media_distribuicao_de_grupos) / desvio_padrao_distribuicao_de_grupos
z_superior = (periodo_medio_superior - media_distribuicao_de_grupos) / desvio_padrao_distribuicao_de_grupos

print(f"\nb. O escore-z para x̄ = {periodo_medio_inferior} minutos é: {z_inferior:.2f}")
print(f"   O escore-z para x̄ = {periodo_medio_superior} minutos é: {z_superior:.2f}")

# c. Encontre a proporção acumulada que corresponda a cada escore-z e calcule a probabilidade.
proporcao_acumulada_inferior = scipy.stats.norm.cdf(z_inferior)
proporcao_acumulada_superior = scipy.stats.norm.cdf(z_superior)

chance_entre_periodos = proporcao_acumulada_superior - proporcao_acumulada_inferior

print(f"\nc. Proporção acumulada para z = {z_inferior:.2f} é: {proporcao_acumulada_inferior:.4f}")
print(f"   Proporção acumulada para z = {z_superior:.2f} é: {proporcao_acumulada_superior:.4f}")
print(f"   A probabilidade de o período médio estar entre {periodo_medio_inferior} e {periodo_medio_superior} minutos é: {chance_entre_periodos:.4f}")

# d. Explique os resultados.
print("\nd. Explicação dos resultados:")
print(f"A probabilidade de que a média do período dispêndido conduzindo por dia por um grupo de {tamanho_do_grupo} condutores (com idade entre 15 e 19 anos) esteja entre {periodo_medio_inferior} e {periodo_medio_superior} minutos é de aproximadamente {chance_entre_periodos:.4f}, ou {chance_entre_periodos*100:.2f}%.")
print("Isso denota que, se você compilasse repetidamente grupos de 100 condutores dessa faixa etária, cerca de 97.45% das vezes a média de tempo dispêndido conduzindo nesses grupos estaria entre 24.7 e 25.5 minutos.")
```

```python
import scipy.stats # Módulo para funcionalidades estatísticas, incluindo a distribuição normal
import matplotlib.pyplot as plt # Módulo para a construção de visualizações gráficas
import numpy as np # Módulo para operações numéricas, como a geração de arranjos

print(" Exercício 18 \n")

# Informações fornecidas
media_total_residencia = 176800  # US$ (µ)
desvio_padrao_total_residencia = 50000  # US$ (σ)
quantidade_no_grupo = 12  # n
valor_medio_desejado = 160000  # US$ (x-barra)

# a. Aplique o teorema do limite central para determinar a média e o desvio padrão da distribuição de grupos de médias.
media_distribuicao_de_grupos = media_total_residencia
desvio_padrao_distribuicao_de_grupos = desvio_padrao_total_residencia / np.sqrt(quantidade_no_grupo)

print(f"a. Média da distribuição de grupos de médias (µ_x̄) = US$ {media_distribuicao_de_grupos:.2f}")
print(f"   Desvio padrão da distribuição de grupos de médias (σ_x̄) = US$ {desvio_padribuicao_de_grupos:.2f}")

# Delineie a distribuição de grupos de médias.
print("\nDelineamento do diagrama da distribuição de grupos de médias:")
print("A distribuição de grupos de médias é uma curva normal, centralizada na média do conjunto total, porém é mais compacta do que a distribuição original do conjunto total.")

# Configuração para o diagrama
eixo_x_grupos = np.linspace(media_distribuicao_de_grupos - 4 * desvio_padrao_distribuicao_de_grupos,
                            media_distribuicao_de_grupos + 4 * desvio_padrao_distribuicao_de_grupos, 500)
eixo_y_grupos = scipy.stats.norm.pdf(eixo_x_grupos, loc=media_distribuicao_de_grupos, scale=desvio_padrao_distribuicao_de_grupos)

plt.figure(figsize=(10, 6))
plt.plot(eixo_x_grupos, eixo_y_grupos, color='blue', label=f'Distribuição de Grupos (n={quantidade_no_grupo}, σ_x̄={desvio_padrao_distribuicao_de_grupos:.2f})')
plt.axvline(media_distribuicao_de_grupos, color='gray', linestyle='--', linewidth=0.8, label=f'Média (µ_x̄) = US$ {media_distribuicao_de_grupos:.2f}')

# Adicionar linha para o valor médio almejado
plt.axvline(valor_medio_desejado, color='red', linestyle='--', label=f'x̄ = US$ {valor_medio_desejado:.2f}')

# b. Calcule o escore-z que corresponda a x-barra = US$ 160.000.
valor_z = (valor_medio_desejado - media_distribuicao_de_grupos) / desvio_padrao_distribuicao_de_grupos
print(f"\nb. O escore-z que corresponde a x̄ = US$ {valor_medio_desejado:.2f} é: {valor_z:.2f}")

# c. Encontre a proporção acumulada que corresponda ao escore-z e calcule a probabilidade.
# A função cdf() retorna a porção à esquerda. Desejamos a porção à direita (maior que).
proporcao_a_esquerda_z = scipy.stats.norm.cdf(valor_z)
chance_maior_que = 1 - proporcao_a_esquerda_z

print(f"\nc. Proporção acumulada para z = {valor_z:.2f} é: {proporcao_a_esquerda_z:.4f}")
print(f"   A probabilidade de o valor médio de comercialização ser maior que US$ {valor_medio_desejado:.2f} é: {chance_maior_que:.4f}")

# Preencher a porção à direita do valor médio almejado no diagrama
eixo_x_preenchimento = np.linspace(valor_medio_desejado, media_distribuicao_de_grupos + 4 * desvio_padrao_distribuicao_de_grupos, 100)
eixo_y_preenchimento = scipy.stats.norm.pdf(eixo_x_preenchimento, loc=media_distribuicao_de_grupos, scale=desvio_padrao_distribuicao_de_grupos)
plt.fill_between(eixo_x_preenchimento, 0, eixo_y_preenchimento, color='salmon', alpha=0.6, label=f'Probabilidade > US$ {valor_medio_desejado:.2f}')

plt.title('Distribuição de Grupos de Médias de Valores de Comercialização de Residências')
plt.xlabel('Valor Médio de Comercialização (US$)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')  # Para evitar notação científica no eixo x
plt.show()

# d. Explique os resultados.
print("\nd. Explicação dos resultados:")
print(f"A probabilidade de que o valor de comercialização médio de um grupo aleatório de {quantidade_no_grupo} moradias seja superior a US$ {valor_medio_desejado:.2f} é de aproximadamente {chance_maior_que:.4f}, ou {chance_maior_que*100:.2f}%.")
print("Isso indica que existe uma elevada chance de que a média de um grupo de 12 moradias ultrapasse US$ 160.000, considerando a média do conjunto total e o desvio padrão.")
```
