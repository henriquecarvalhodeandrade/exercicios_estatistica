```python
import numpy as np
import scipy.stats

print("Exercício 37\n")

# Dados fornecidos
tamanho_grupo = 30           # Dimensão da amostra (n)
desvio_padrao_grupo = 1.20   # Desvio padrão do grupo (miligramas)
variancia_grupo = desvio_padrao_grupo**2 # Variância do grupo

niveis_de_confianca_desejados = [0.90, 0.95] # Níveis de confiança a serem calculados

# Graus de liberdade
graus_de_liberdade = tamanho_grupo - 1
print(f"Graus de liberdade (df) = n - 1 = {graus_de_liberdade}")
print(f"Variância do grupo (s²) = {variancia_grupo:.2f} (miligramas²)")

for nivel_confianca in niveis_de_confianca_desejados:
    print(f"\n--- Intervalo de Confiança de {nivel_confianca*100:.0f}% ---")

    # a. Encontre os valores críticos de χ²_R e χ²_L para cada intervalo de confiança.
    # A área nas caudas é (1 - nível_confianca). Cada cauda tem metade dessa área.
    alfa_dividido_por_2 = (1 - nivel_confianca) / 2

    chi2_esquerda = scipy.stats.chi2.ppf(alfa_dividido_por_2, df=graus_de_liberdade)
    chi2_direita = scipy.stats.chi2.ppf(1 - alfa_dividido_por_2, df=graus_de_liberdade)

    print(f"a. Valores críticos Qui-Quadrado para df={graus_de_liberdade}:")
    print(f"   χ²_L (cauda esquerda, área à esquerda={alfa_dividido_por_2:.3f}) = {chi2_esquerda:.3f}")
    print(f"   χ²_R (cauda direita, área à esquerda={1-alfa_dividido_por_2:.3f}) = {chi2_direita:.3f}")

    # b. Calcule os limites inferior e superior para a variância populacional.
    limite_inferior_variancia = ((tamanho_grupo - 1) * variancia_grupo) / chi2_direita
    limite_superior_variancia = ((tamanho_grupo - 1) * variancia_grupo) / chi2_esquerda

    print(f"\nb. Intervalo de Confiança para a **VARIÂNCIA Populacional (σ²)**:")
    print(f"   Limite Inferior = {limite_inferior_variancia:.3f} miligramas²")
    print(f"   Limite Superior = {limite_superior_variancia:.3f} miligramas²")

    # c. Calcule as raízes quadradas dos limites para o desvio padrão.
    limite_inferior_desvio_padrao = np.sqrt(limite_inferior_variancia)
    limite_superior_desvio_padrao = np.sqrt(limite_superior_variancia)

    # d. Especifique os intervalos de confiança.
    print(f"\nd. Intervalo de Confiança de {nivel_confianca*100:.0f}% para a **VARIÂNCIA Populacional (σ²)**:")
    print(f"   ({limite_inferior_variancia:.3f}, {limite_superior_variancia:.3f}) miligramas²")

    print(f"   Intervalo de Confiança de {nivel_confianca*100:.0f}% para o **DESVIO PADRÃO Populacional (σ)**:")
    print(f"   ({limite_inferior_desvio_padrao:.3f}, {limite_superior_desvio_padrao:.3f}) miligramas")

    print(f"\n**Interpretação** para {nivel_confianca*100:.0f}% de confiança:")
    print(f"Estamos {nivel_confianca*100:.0f}% confiantes de que a verdadeira variância populacional dos pesos dos remédios está entre {limite_inferior_variancia:.3f} e {limite_superior_variancia:.3f} miligramas².")
    print(f"E estamos {nivel_confianca*100:.0f}% confiantes de que o verdadeiro desvio padrão populacional está entre {limite_inferior_desvio_padrao:.3f} e {limite_superior_desvio_padrao:.3f} miligramas.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

print("Exercício 40\n")

def plotar_area_p_valor(valor_z, tipo_teste, sufixo_titulo=""):
    """Delineia uma distribuição normal padrão e preenche a área para o valor p."""
    eixo_x = np.linspace(-4, 4, 500)
    eixo_y = scipy.stats.norm.pdf(eixo_x)

    plt.figure(figsize=(8, 5))
    plt.plot(eixo_x, eixo_y, color='blue')
    plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)')

    if tipo_teste == "unilateral à esquerda":
        x_preencher = np.linspace(-4, valor_z, 100)
        y_preencher = scipy.stats.norm.pdf(x_preencher)
        plt.fill_between(x_preencher, 0, y_preencher, color='skyblue', alpha=0.6, label='Área p-valor')
        plt.axvline(valor_z, color='red', linestyle='--', label=f'z = {valor_z:.2f}')
        plt.title(f'Teste Unilateral à Esquerda {sufixo_titulo}')
    elif tipo_teste == "unilateral à direita":
        x_preencher = np.linspace(valor_z, 4, 100)
        y_preencher = scipy.stats.norm.pdf(x_preencher)
        plt.fill_between(x_preencher, 0, y_preencher, color='salmon', alpha=0.6, label='Área p-valor')
        plt.axvline(valor_z, color='red', linestyle='--', label=f'z = {valor_z:.2f}')
        plt.title(f'Teste Unilateral à Direita {sufixo_titulo}')
    elif tipo_teste == "bilateral":
        z_absoluto = abs(valor_z)
        x_preencher_esquerda = np.linspace(-4, -z_absoluto, 100)
        y_preencher_esquerda = scipy.stats.norm.pdf(x_preencher_esquerda)
        plt.fill_between(x_preencher_esquerda, 0, y_preencher_esquerda, color='lightgreen', alpha=0.6, label='Área p-valor')

        x_preencher_direita = np.linspace(z_absoluto, 4, 100)
        y_preencher_direita = scipy.stats.norm.pdf(x_preencher_direita)
        plt.fill_between(x_preencher_direita, 0, y_preencher_direita, color='lightgreen', alpha=0.6)

        plt.axvline(-z_absoluto, color='red', linestyle='--', label=f'z = -{z_absoluto:.2f}')
        plt.axvline(z_absoluto, color='red', linestyle='--', label=f'z = {z_absoluto:.2f}')
        plt.title(f'Teste Bilateral {sufixo_titulo}')
    else:
        plt.title('Distribuição Normal Padrão') # Fallback para o título, caso o tipo de teste não seja reconhecido
        
    plt.xlabel('Z-score')
    plt.ylabel('Densidade de Probabilidade')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.show()

# --- Declaração 1 ---
print("\n1. Um analista de consumo informa que a durabilidade média de um tipo específico de bateria automotiva não é de 74 meses.")
# a. Expresse H₀ e Hₐ em palavras e em símbolos.
H0_simbolo_1 = "μ = 74"
Ha_simbolo_1 = "μ ≠ 74"
print(f"a. H₀ (palavras): A durabilidade média da bateria é de 74 meses. (Símbolo: {H0_simbolo_1})")
print(f"   Hₐ (palavras): A durabilidade média da bateria NÃO é de 74 meses. (Símbolo: {Ha_simbolo_1})")
# b. Determine se o teste é unilateral à esquerda, unilateral à direita ou bilateral.
tipo_teste_1 = "bilateral"
print(f"b. O teste é: **{tipo_teste_1}** (pois Hₐ contém ≠)")
# c. Delineie a distribuição normal padrão e preencha a área para o valor p.
# O valor p para um teste bilateral é 2 * P(Z > |z_estatistica|) ou 2 * P(Z < -|z_estatistica|).
# Para preencher, usamos |z_estatistica| (vamos usar um exemplo de z-estatística, por exemplo, 1.5, para o plot)
plotar_area_p_valor(1.5, tipo_teste_1, " (μ ≠ 74)")

# --- Declaração 2 ---
print("\n2. Um agente imobiliário alega que a proporção de proprietários que consideram suas residências muito pequenas para suas famílias é superior a 24%.")
# a. Expresse H₀ e Hₐ em palavras e em símbolos.
H0_simbolo_2 = "p ≤ 0.24"
Ha_simbolo_2 = "p > 0.24"
print(f"a. H₀ (palavras): A proporção de proprietários é menor ou igual a 24%. (Símbolo: {H0_simbolo_2})")
print(f"   Hₐ (palavras): A proporção de proprietários é superior a 24%. (Símbolo: {Ha_simbolo_2})")
# b. Determine se o teste é unilateral à esquerda, unilateral à direita ou bilateral.
tipo_teste_2 = "unilateral à direita"
print(f"b. O teste é: **{tipo_teste_2}** (pois Hₐ contém >)")
# c. Delineie a distribuição normal padrão e preencha a área para o valor p.
# Para preencher, usamos um exemplo de z-estatística, por exemplo, 1.0, para o plot
plotar_area_p_valor(1.0, tipo_teste_2, " (p > 0.24)")
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("Exercício 49 - Teste 1: Salário dos Engenheiros Mecânicos\n")

# Informações para a análise
media_hipotese_salario = 68000  # US$ (Média hipotética da H₀)
tamanho_grupo_salario = 20      # Dimensão do grupo
media_grupo_salario = 66900     # US$ (Média observada no grupo)
desvio_padrao_populacao_salario = 5500  # US$ (Desvio padrão conhecido da população)
nivel_significancia_salario = 0.05      # Nível de significância (α)

# a. Formulação das hipóteses
print("a. Formulação das Hipóteses:")
print(f"   H₀: μ ≥ {media_hipotese_salario} (Salário médio é maior ou igual a US$ {media_hipotese_salario})")
print(f"   Hₐ: μ < {media_hipotese_salario} (Salário médio é menor que US$ {media_hipotese_salario} - ALEGAÇÃO DOS FUNCIONÁRIOS)")
print("   Tipo de Teste: Unilateral à esquerda")

# b. Nível de significância
print(f"\nb. Nível de Significância (α) = {nivel_significancia_salario:.2f}")

# c. Encontre o valor crítico z_c.
# Para um teste unilateral à esquerda, z_c é o valor que deixa α de área à sua esquerda.
z_critico_salario = scipy.stats.norm.ppf(nivel_significancia_salario)
print(f"\nc. Valor crítico z_c = {z_critico_salario:.2f}")

# d. Calcule a estatística de teste padronizada z.
erro_padrao_salario = desvio_padrao_populacao_salario / np.sqrt(tamanho_grupo_salario)
z_estatistico_salario = (media_grupo_salario - media_hipotese_salario) / erro_padrao_salario
print(f"   Estatística de teste z = {z_estatistico_salario:.2f}")

# e. Decisão
print("\ne. Decisão (abordagem do valor crítico):")
if z_estatistico_salario < z_critico_salario:
    decisao_salario = "Rejeitar H₀"
    interpretacao_salario = f"Há **evidência estatística suficiente** (com α = {nivel_significancia_salario:.2f}) para **apoiar a alegação dos funcionários** de que o salário médio dos engenheiros mecânicos é menor que US$ {media_hipotese_salario}."
else:
    decisao_salario = "Não rejeitar H₀"
    interpretacao_salario = f"Não há **evidência estatística suficiente** (com α = {nivel_significancia_salario:.2f}) para **apoiar a alegação dos funcionários** de que o salário médio dos engenheiros mecânicos é menor que US$ {media_hipotese_salario}."

print(f"   Decisão: {decisao_salario}")
print(f"   Interpretação: {interpretacao_salario}")

# Gráfico para o Teste 1 (Salário)
x_valores_salario = np.linspace(media_hipotese_salario - 3 * erro_padrao_salario, media_hipotese_salario + 3 * erro_padrao_salario, 500)
y_valores_salario = scipy.stats.norm.pdf(x_valores_salario, loc=media_hipotese_salario, scale=erro_padrao_salario)

plt.figure(figsize=(10, 6))
plt.plot(x_valores_salario, y_valores_salario, color='blue', label='Distribuição Amostral de Salários')
plt.axvline(media_hipotese_salario, color='gray', linestyle='--', linewidth=0.8, label=f'Média H₀ = ${media_hipotese_salario}')
plt.axvline(z_critico_salario * erro_padrao_salario + media_hipotese_salario, color='red', linestyle='--', label=f'Limite Crítico z_c = {z_critico_salario:.2f}')
plt.axvline(media_grupo_salario, color='green', linestyle='-', label=f'Média Amostral = ${media_grupo_salario}')

x_regiao_rejeicao_salario = np.linspace(x_valores_salario.min(), z_critico_salario * erro_padrao_salario + media_hipotese_salario, 100)
y_regiao_rejeicao_salario = scipy.stats.norm.pdf(x_regiao_rejeicao_salario, loc=media_hipotese_salario, scale=erro_padrao_salario)
plt.fill_between(x_regiao_rejeicao_salario, 0, y_regiao_rejeicao_salario, color='salmon', alpha=0.6, label=f'Região de Rejeição (α={nivel_significancia_salario})')

plt.title('Teste de Hipótese para Salário de Engenheiros (Unilateral à Esquerda)')
plt.xlabel('Salário (US$)')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np

print("Exercício 53\n")

# Informações para a análise
custo_medio_hipotetico = 1200  # US$ (Média hipotética da H₀)
tamanho_grupo = 7              # Dimensão do grupo
custo_medio_observado = 1125   # US$ (Média observada no grupo)
desvio_padrao_grupo = 55       # US$ (Desvio padrão do grupo)
nivel_significancia = 0.10     # Nível de significância (α)
populacao_eh_normal = True     # Assumido no problema

# a. Identifique a alegação e formule H₀ e Hₐ.
alegacao_53 = "O custo médio do seguro do sedan de dois anos é menor que US$ 1.200 (μ < 1200)."
print(f"a. Alegação: {alegacao_53}")

H0_simbolo = "μ ≥ 1200"
Ha_simbolo = "μ < 1200" # A alegação do agente
print(f"   H₀: {H0_simbolo} (Hipótese Nula)")
print(f"   Hₐ: {Ha_simbolo} (Hipótese Alternativa - REPRESENTA A ALEGAÇÃO)")
print("   Tipo de Teste: Unilateral à esquerda")

# b. Identifique o nível de significância α e os graus de liberdade.
graus_de_liberdade = tamanho_grupo - 1 # Graus de liberdade
print(f"\nb. Nível de Significância (α) = {nivel_significancia:.2f}")
print(f"   Graus de Liberdade (df) = n - 1 = {graus_de_liberdade}")

# c. Encontre o valor crítico t₀ e identifique a região de rejeição.
# Para um teste unilateral à esquerda, t₀ é o valor t que deixa α de área à sua esquerda.
t_critico = scipy.stats.t.ppf(nivel_significancia, df=graus_de_liberdade)
print(f"\nc. Valor crítico t₀ = {t_critico:.3f}")
print(f"   Região de Rejeição: Rejeitar H₀ se a estatística de teste t < **{t_critico:.3f}**.")

# d. Calcule a estatística de teste padronizada t. Delineie um diagrama.
erro_padrao_amostral = desvio_padrao_grupo / np.sqrt(tamanho_grupo) # s / sqrt(n) é o erro padrão do grupo
t_estatistico = (custo_medio_observado - custo_medio_hipotetico) / erro_padrao_amostral
print(f"\nd. Estatística de teste padronizada t = {t_estatistico:.2f}")

# Delineamento do diagrama da distribuição t
valores_x_t = np.linspace(-4, 4, 500) # Eixo x para a distribuição t
valores_y_t = scipy.stats.t.pdf(valores_x_t, df=graus_de_liberdade) # PDF da distribuição t

plt.figure(figsize=(10, 6)) # Define o tamanho da figura
plt.plot(valores_x_t, valores_y_t, color='blue', label=f'Distribuição t (df={graus_de_liberdade})') # Plota a curva t
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8, label='Média (0)') # Linha da média
plt.axvline(t_critico, color='red', linestyle='--', label=f't₀ = {t_critico:.3f}') # Linha do valor crítico t0
plt.axvline(t_estatistico, color='green', linestyle=':', label=f'Estatística de Teste t = {t_estatistico:.2f}') # Linha da estatística de teste

# Preencher a região de rejeição (cauda esquerda)
x_preenchimento_esquerda = np.linspace(-4, t_critico, 100)
y_preenchimento_esquerda = scipy.stats.t.pdf(x_preenchimento_esquerda, df=graus_de_liberdade)
plt.fill_between(x_preenchimento_esquerda, 0, y_preenchimento_esquerda, color='salmon', alpha=0.6, label=f'Região de Rejeição (α={nivel_significancia:.2f})')

plt.title(f'Teste de Hipótese para Média (Unilateral à Esquerda, α = {nivel_significancia:.2f})') # Título do gráfico
plt.xlabel('Estatística t') # Rótulo do eixo X
plt.ylabel('Densidade de Probabilidade') # Rótulo do eixo Y
plt.legend() # Mostra a legenda
plt.grid(True, linestyle=':', alpha=0.7) # Adiciona grade
plt.show() # Exibe o gráfico

# e. Decida se a hipótese nula deve ser rejeitada.
print(f"\ne. Decisão:")
if t_estatistico < t_critico: # Se a estatística de teste cair na região de rejeição
    decisao_final = "Rejeitar H₀"
else:
    decisao_final = "Não rejeitar H₀"
print(f"   Decisão: {decisao_final}")

# f. Interprete a decisão no contexto da alegação original.
print(f"\nf. Interpretação da decisão:")
if decisao_final == "Rejeitar H₀":
    print(f"   Há **evidência estatística suficiente** (com α = {nivel_significancia:.2f}) para **concordar com a alegação do agente** de que o custo médio do seguro do sedan de dois anos é menor que US$ {custo_medio_hipotetico}.")
else:
    print(f"   Não há **evidência estatística suficiente** (com α = {nivel_significancia:.2f}) para **concordar com a alegação do agente** de que o custo médio do seguro do sedan de dois anos é menor que US$ {custo_medio_hipotetico}.")
```
